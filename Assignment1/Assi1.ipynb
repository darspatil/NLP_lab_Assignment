```python
import nltk
from nltk.tokenize import (
    word_tokenize,
    TreebankWordTokenizer,
    TweetTokenizer,
    MWETokenizer
)
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
text = "i go to Mumbai and i watched the gatway of India, marine drive after come home i did study about dsa and logic building, an builds projects around full stack and gen ai, also worked in sugarcane farm with my parents. Also in daily i go for tracking"
```

```python
whitespace_tokens = text.split()
```

```python
punct_tokens = word_tokenize(text)
```

```python
treebank_tokenizer = TreebankWordTokenizer()
treebank_tokens = treebank_tokenizer.tokenize(text)
```

```python
tweet_tokenizer = TweetTokenizer()
tweet_tokens = tweet_tokenizer.tokenize(text)
```

```python
mwe_tokenizer = MWETokenizer([('gatway', 'of', 'India',), ('full', 'stack',), ('gen', 'ai',)])
mwe_tokens = mwe_tokenizer.tokenize(punct_tokens)
```

```python
print(f"Whitespace: {whitespace_tokens[8:11], whitespace_tokens[28:35]}")
print(f"Treebank:   {treebank_tokens}")
print(f"Tweet:      {tweet_tokens}")
print(f"MWE:        {mwe_tokens[8:11], mwe_tokens[28:35]}")
```

```python
from nltk.stem import PorterStemmer, SnowballStemmer, WordNetLemmatizer
MWEstemmer = PorterStemmer()
SBstemmer = SnowballStemmer("english")
```

```python
MWEstems = [MWEstemmer.stem(mw) for mw in tweet_tokens]
print(MWEstems)
```

```python
SBstems = [SBstemmer.stem(mw) for mw in mwe_tokens]
print(MWEstems)
```

```python
import nltk
nltk.download('averaged_perceptron_tagger_eng')

pos_tags = nltk.pos_tag(mwe_tokens)
print(pos_tags)
```

```python
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

def to_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
lemmatized = [
    lemmatizer.lemmatize(word, to_wordnet_pos(tag))
    for word, tag in pos_tags
]

print(lemmatized)
```
