```python
from transformers import pipeline
model_path = "cardiffnlp/twitter-xlm-roberta-base-sentiment"
sentiment_analyzer = pipeline("sentiment-analysis", model=model_path, tokenizer=model_path)

print("Model successfully loaded!")
```

```python
hindi_sentences = [
    # Translation: "This movie is very brilliant and I liked it a lot."
    "यह फिल्म बहुत शानदार है और मुझे यह बहुत पसंद आई।",

    # Translation: "The customer service was very bad and my experience was disappointing."
    "ग्राहक सेवा बहुत खराब थी और मेरा अनुभव निराशाजनक रहा।",

    # Translation: "I will go to the market tomorrow and buy some vegetables."
    "मैं कल बाजार जाकर कुछ सब्जियां खरीदूंगा।"
]
```

```python
print("--- Hindi Sentiment Analysis Results ---\n")

for sentence in hindi_sentences:
    result = sentiment_analyzer(sentence)[0]
    label = result['label'].upper()
    confidence = result['score']

    print(f"Text: {sentence}")
    print(f"Predicted Sentiment: {label}")
    print(f"Confidence Score:  {confidence:.2%}\n")
```
